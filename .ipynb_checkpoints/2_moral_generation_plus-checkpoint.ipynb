# Moral Generation + Narrative Dimensions

For every book, this notebook calls one or more LLMs with the prompt in
`prompt.txt` to extract, in a single structured-JSON pass:

- the **story morals** (exactly three), **and**
- a set of **narrative dimensions** — protagonist, themes, locations, central
  conflict, principal actions, values pursued, values threatened, and
  consequence/resolution.

It runs across two axes:

- **Models** (default **GPT + Gemini**) — compare how models read the same book.
- **Input representations** (`CONDITIONS`):
  - `full_text` — the raw book text
  - `chunk_summary` — the aggregated chunk summary from `summaries.csv`
  - `memory` — **no text supplied**; the model uses its own knowledge of the
    work, identified by *Title* and *Author* from `meta.csv`

The base prompt is **adjusted per condition** (the opening framing and the
trailing input block change; all instructions and the JSON schema stay intact).

**This notebook does NOT assign value labels** — that is a separate stage.

### Output
`moral_generation_plus.csv`, one row per (`filename` x `input_condition` x
`model`):

`filename | input_condition | model | output_json | story_moral_1..3 | parse_ok | timestamp`

`output_json` holds the **full** structured result (all narrative dimensions);
the three morals are broken out for convenience. Parse narrative dimensions from
`output_json` downstream.

### How to use
Edit the **Parameters** cell, then run top to bottom. The notebook checkpoints
after each generation and skips (book, condition, model) combos already done.

## 1. Imports

````python
import re, json, time
from getpass import getpass
from pathlib import Path
from datetime import datetime, timezone

import pandas as pd
from openai import OpenAI
````

## 2. Parameters — edit these

````python
# --- Paths -------------------------------------------------------------------
INPUT_DIR     = Path("InputTexts")              # full-text .txt books (full_text)
SUMMARIES_CSV = Path("summaries.csv")           # chunk summaries (chunk_summary)
META_CSV      = Path("meta.csv")                # Title/Author (memory)
PROMPT_TXT    = Path("prompt.txt")              # the moral + narrative prompt
OUTPUT_CSV    = Path("moral_generation_plus.csv")

# --- Models ------------------------------------------------------------------
# Default GPT + Gemini. Add/remove models here (provider inferred from prefix).
MODELS = [
    "gpt-5.4",                  # comment out a line to skip that model
    "gemini-3.1-pro-preview",
]

# --- Input representations to run --------------------------------------------
#   full_text     : the raw book text
#   chunk_summary : the aggregated chunk summary from summaries.csv
#   memory        : no text; model uses its knowledge of <Title> by <Author>
CONDITIONS = [
    "full_text",        # raw book text  (comment out this line to run WITHOUT full text)
    "chunk_summary",    # aggregated chunk summary from summaries.csv
    "memory",           # no text; model uses its knowledge of <Title> by <Author>
]

# --- Generation settings -----------------------------------------------------
TEMPERATURE       = 0
USE_TEMPERATURE   = True     # set False for reasoning models that reject `temperature`
MAX_OUTPUT_TOKENS = 2000     # the structured JSON output can be sizable

# --- Robustness / rate limiting ----------------------------------------------
DELAY_SECONDS   = 0.1
MAX_RETRIES     = 5
REQUEST_TIMEOUT = 120
SAVE_EVERY      = 1          # checkpoint after every N generations
````

## 3. Prompt + per-condition adaptation

We load `prompt.txt`, split off the trailing `Story: [INSERT STORY HERE]`
placeholder, and swap in a condition-appropriate opening sentence and input
block. All numbered instructions and the JSON schema are preserved verbatim.

````python
PROMPT_TEXT = (PROMPT_TXT.read_text(encoding="utf-8")
               .replace("\u2028", "\n").replace("\u2029", "\n"))

# Drop the trailing input placeholder ("Story: [INSERT STORY HERE]") so we
# can swap in a condition-appropriate opening + input. The regex tolerates
# any whitespace (including newlines) between "Story:" and the bracket.
_body = re.split(r"Story:\s*\[INSERT STORY HERE\]", PROMPT_TEXT)[0].rstrip()

# The body opens with this fixed framing sentence; strip it to replace it.
_LEAD = "Given the following story,"
assert _body.startswith(_LEAD), "prompt.txt opening changed; update _LEAD"
INSTRUCTIONS = _body[len(_LEAD):].lstrip()   # text after the opening phrase


def build_prompt(condition, text=None, title=None, author=None):
    # Adjust the opening framing + trailing input for each representation.
    if condition == "full_text":
        opening = "Given the following story,"
        tail = f"\n\nStory:\n{text}"
    elif condition == "chunk_summary":
        opening = "Given the following plot summary of a story,"
        tail = f"\n\nPlot Summary:\n{text}"
    elif condition == "memory":
        # No text supplied -- rely on the model's own knowledge of the work.
        opening = f'For the novel "{title}" by {author},'
        tail = ""
    else:
        raise ValueError(f"Unknown condition: {condition}")
    return f"{opening} {INSTRUCTIONS}{tail}"

# Preview the memory-condition prompt opening as a sanity check.
print(build_prompt("memory", title="<Title>", author="<Author>")[:160], "...")
````

## 4. API keys + model calls

Keys are requested live (hidden prompt) — never hard-coded or saved. A Gemini
key is only requested if a `gemini-*` model is in `MODELS`.

````python
openai_key = getpass("Enter your OpenAI API key: ")
openai_client = OpenAI(api_key=openai_key)

gemini_key = None
if any(m.lower().startswith("gemini") for m in MODELS):
    gemini_key = getpass("Enter your Gemini (Google) API key: ")


def provider_for_model(model):
    m = model.lower()
    if m.startswith(("gpt", "o1", "o3")): return "openai"
    if m.startswith("gemini"):            return "google"
    raise ValueError(f"Unsupported model: {model}")


def _call_openai(model, prompt):
    # response_format=json_object asks the model for strict JSON.
    kwargs = dict(model=model,
                  messages=[{"role": "user", "content": prompt}],
                  response_format={"type": "json_object"},
                  max_completion_tokens=MAX_OUTPUT_TOKENS,
                  timeout=REQUEST_TIMEOUT)
    if USE_TEMPERATURE:
        kwargs["temperature"] = TEMPERATURE
    resp = openai_client.chat.completions.create(**kwargs)
    return resp.choices[0].message.content


def _call_gemini(model, prompt):
    import requests
    gen = {"responseMimeType": "application/json"}
    if USE_TEMPERATURE:
        gen["temperature"] = TEMPERATURE
    r = requests.post(
        f"https://generativelanguage.googleapis.com/v1beta/models/{model}:generateContent",
        headers={"x-goog-api-key": gemini_key, "content-type": "application/json"},
        json={"contents": [{"parts": [{"text": prompt}]}], "generationConfig": gen},
        timeout=REQUEST_TIMEOUT)
    if r.status_code != 200:
        raise RuntimeError(f"Gemini {r.status_code}: {r.text[:300]}")
    data = r.json()
    cands = data.get("candidates") or []
    if not cands:
        raise RuntimeError(f"Gemini: no candidates ({json.dumps(data)[:300]})")
    cand = cands[0]
    parts = (cand.get("content") or {}).get("parts") or []
    # Join all real (non-thought) text parts; reasoning models may emit several.
    texts = [p["text"] for p in parts
             if isinstance(p, dict) and "text" in p and not p.get("thought")]
    if not texts:
        raise RuntimeError("Gemini: empty reply "
                           f"(finishReason={cand.get('finishReason')}; raw={json.dumps(data)[:300]})")
    return "".join(texts)


def call_model(model, prompt):
    # Dispatch by provider, retrying transient (429 / 5xx / timeout) errors.
    provider = provider_for_model(model)
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            if provider == "openai": return _call_openai(model, prompt)
            if provider == "google": return _call_gemini(model, prompt)
        except Exception as e:
            transient = any(s in str(e).lower() for s in
                            ("429", "500", "502", "503", "529", "timeout", "overloaded"))
            if attempt == MAX_RETRIES or not transient:
                raise
            wait = 2 ** attempt
            print(f"      {model} error ({e}); retry in {wait}s ({attempt}/{MAX_RETRIES})")
            time.sleep(wait)
````

## 5. Load metadata + summaries, and helpers

````python
def normalize_id(name):
    # Canonical book id = filename without a trailing .txt
    name = str(name).strip()
    return name[:-4] if name.lower().endswith(".txt") else name


def read_text_file(path):
    try:
        return Path(path).read_text(encoding="utf-8")
    except UnicodeDecodeError:
        return Path(path).read_text(encoding="utf-8", errors="replace")


def extract_json(txt):
    # Parse a JSON object from the model reply (tolerant of code fences / prose).
    if not txt:
        return None
    t = txt.strip()
    if t.startswith("```"):
        t = re.sub(r"^```(?:json)?", "", t).rsplit("```", 1)[0].strip()
    try:
        return json.loads(t)
    except Exception:
        m = re.search(r"\{.*\}", t, re.S)   # fall back to the outermost {...}
        if m:
            try:
                return json.loads(m.group(0))
            except Exception:
                return None
        return None


# Title/Author keyed by canonical id — used by the memory condition.
meta_df = pd.read_csv(META_CSV)
meta = {}
for _, row in meta_df.iterrows():
    meta[normalize_id(row["filename"])] = {"title": row.get("Title"),
                                           "author": row.get("Author")}

# Chunk summaries keyed by canonical id — used by the chunk_summary condition.
summaries = {}
if SUMMARIES_CSV.exists():
    sdf = pd.read_csv(SUMMARIES_CSV)
    for _, row in sdf.iterrows():
        summaries[normalize_id(row["filename"])] = row["chunk_summary"]
else:
    print(f"WARNING: {SUMMARIES_CSV} not found — chunk_summary condition will be skipped.")

print(f"{len(meta)} book(s) in metadata; {len(summaries)} chunk summary(ies) loaded.")
````

## 6. Main loop - books x conditions x models

Writes each generation to the canonical **`moral_generation_plus.jsonl`** (full structured result, the checkpoint) and rebuilds a slim **`moral_generation_plus.csv`** index. Skips combos already in the JSONL.

````python
RAW_JSONL  = OUTPUT_CSV.with_suffix(".jsonl")   # canonical, lossless raw store
INDEX_COLS = ["filename", "input_condition", "model", "parse_ok",
              "story_moral_1", "story_moral_2", "story_moral_3", "timestamp"]

def _key(filename, condition, model):
    return (str(filename), str(condition), str(model))

def moral_at(morals, i):
    return morals[i] if isinstance(morals, list) and len(morals) > i else None

# Resume: a (book, condition, model) is done if already present in the JSONL.
done = set()
if RAW_JSONL.exists():
    with open(RAW_JSONL, encoding="utf-8") as f:
        for line in f:
            if line.strip():
                d = json.loads(line)
                done.add(_key(d["filename"], d["input_condition"], d["model"]))
    print(f"Resuming - {len(done)} generation(s) already in {RAW_JSONL}")

n_new = 0
for book_id, info in meta.items():
    for condition in CONDITIONS:
        # Resolve the input for this condition (skip if unavailable).
        text = None
        if condition == "full_text":
            p = INPUT_DIR / f"{book_id}.txt"
            if not p.exists():
                print(f"SKIP  {book_id} / full_text (no .txt found)")
                continue
            text = read_text_file(p)
        elif condition == "chunk_summary":
            text = summaries.get(book_id)
            if not text:
                print(f"SKIP  {book_id} / chunk_summary (not in summaries.csv)")
                continue
        # memory: no text needed (uses title/author from meta.csv)

        for model in MODELS:
            if _key(book_id, condition, model) in done:
                print(f"SKIP  {book_id} / {condition} / {model} (done)")
                continue
            print(f"GEN   {book_id} / {condition} / {model}")
            try:
                prompt = build_prompt(condition, text=text,
                                      title=info["title"], author=info["author"])
                raw = call_model(model, prompt)
                parsed = extract_json(raw)
                record = {
                    "filename": book_id,
                    "input_condition": condition,
                    "model": model,
                    "parse_ok": parsed is not None,
                    "timestamp": datetime.now(timezone.utc).isoformat(timespec="seconds"),
                    "result": parsed,                 # full structured object (None on parse fail)
                    "raw_text": None if parsed is not None else raw,  # keep raw reply on failure
                }
                # Append to the canonical JSONL immediately (this is the checkpoint).
                with open(RAW_JSONL, "a", encoding="utf-8") as f:
                    f.write(json.dumps(record, ensure_ascii=False) + "\n")
                done.add(_key(book_id, condition, model))
                n_new += 1
                time.sleep(DELAY_SECONDS)
            except Exception as e:
                print(f"      ERROR: {e}")

# Rebuild the slim CSV index from the canonical JSONL (single source of truth).
rows = []
if RAW_JSONL.exists():
    with open(RAW_JSONL, encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue
            d = json.loads(line)
            res = d.get("result")
            morals = res.get("story_morals") if isinstance(res, dict) else None
            rows.append({
                "filename": d["filename"], "input_condition": d["input_condition"],
                "model": d["model"], "parse_ok": d["parse_ok"],
                "story_moral_1": moral_at(morals, 0), "story_moral_2": moral_at(morals, 1),
                "story_moral_3": moral_at(morals, 2), "timestamp": d["timestamp"],
            })
results = pd.DataFrame(rows, columns=INDEX_COLS)
results.to_csv(OUTPUT_CSV, index=False)
print(f"\nDone. {len(results)} generation(s).")
print(f"  raw   -> {RAW_JSONL}")
print(f"  index -> {OUTPUT_CSV}")
print("Now run 3_parse_outputs.ipynb to build the tidy tables.")
````
