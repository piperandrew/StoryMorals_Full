# 5. Validation table (one story, all conditions & models)

Flattens the structured outputs for a single story into a long table for
eyeballing every model answer. By default it includes **all input conditions
and all models** for that story, so each row is identified by its generation:

`filename | input_condition | model | main_category | sub_category | answer`

- **main_category** = top-level field (`protagonist`, `major_themes`,
  `central_conflict`, `story_morals`, ...).
- **sub_category** = the path within it (a sub-field like `name`; an array index
  like `[1]`; or both, e.g. `[1].location` for a list of objects).
- **answer** = the model's value at that leaf.

One row per leaf, per generation.

- **Input:** `moral_generation_plus.jsonl` (from 2_moral_generation_plus)
- **Output:** `validation_table.csv`

By default it validates the **first story** in the file (all its conditions /
models). Set `STORY` to a filename to pick another; narrow `CONDITIONS` / `MODELS`
to a subset if you want.

## 1. Imports + parameters

```python
import json
from pathlib import Path
import pandas as pd

RAW_JSONL  = Path("moral_generation_plus.jsonl")   # stage-2 output
OUTPUT_CSV = Path("validation_table.csv")

# Which story to validate. None -> the first story (filename) in the file.
STORY = None
# Restrict to a subset if you like; None means ALL conditions / ALL models.
CONDITIONS = None     # e.g. ["full_text", "chunk_summary", "memory"]
MODELS     = None     # e.g. ["gpt-5.4", "gemini-3.1-pro-preview"]
```

## 2. Load and select the story's generations

```python
records = []
with open(RAW_JSONL, encoding="utf-8") as f:
    for line in f:
        if line.strip():
            records.append(json.loads(line))
if not records:
    raise ValueError(f"No records in {RAW_JSONL}")

story = STORY if STORY is not None else records[0]["filename"]

selected = [r for r in records
            if r["filename"] == story
            and (CONDITIONS is None or r["input_condition"] in CONDITIONS)
            and (MODELS is None or r["model"] in MODELS)]
if not selected:
    raise ValueError(f"No records match story={story!r} with the given filters.")

print(f"Story: {story!r}")
print(f"{len(selected)} generation(s):")
for r in selected:
    print(f"  - {r['input_condition']} / {r['model']} (parse_ok={r.get('parse_ok')})")
```

## 3. Flatten each generation to main_category / sub_category / answer

`walk()` recurses through objects and arrays, emitting one `(sub_path, value)`
pair per leaf. Array items get a `[i]` index; nested object fields are
dot-joined (e.g. `[1].location`).

```python
def walk(val, prefix=""):
    out = []
    if isinstance(val, dict):
        for k, v in val.items():
            sub = k if not prefix else f"{prefix}.{k}"
            out.extend(walk(v, sub))
    elif isinstance(val, list):
        for i, item in enumerate(val, 1):
            sub = f"{prefix}[{i}]" if prefix else f"[{i}]"
            out.extend(walk(item, sub))
    else:
        out.append((prefix, val))
    return out


rows = []
for rec in selected:
    base = {"filename": rec["filename"],
            "input_condition": rec["input_condition"],
            "model": rec["model"]}
    result = rec.get("result")
    if not isinstance(result, dict):
        # Keep a marker row so an unparsed generation is visible, not silently gone.
        rows.append({**base, "main_category": None, "sub_category": None,
                     "answer": "<no parsed result (parse_ok=False)>"})
        continue
    for main, val in result.items():
        leaves = walk(val)
        if not leaves:                   # empty object/array -> show the category
            rows.append({**base, "main_category": main, "sub_category": "", "answer": None})
        for sub, ans in leaves:
            rows.append({**base, "main_category": main, "sub_category": sub, "answer": ans})

vt = pd.DataFrame(rows, columns=["filename", "input_condition", "model",
                                 "main_category", "sub_category", "answer"])
vt.to_csv(OUTPUT_CSV, index=False)
print(f"{len(vt)} output rows -> {OUTPUT_CSV}")
```
